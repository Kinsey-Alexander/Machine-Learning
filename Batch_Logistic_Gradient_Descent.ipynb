```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = datasets.load_iris()

# Create a Pandas DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names) # importing feature matrix
iris_df['target'] = iris.target # importing target array; 0 = iris setosa, 1 = iris virginica, 2 = iris versicolor


# Simplifying dataset for purposes of this exercise:
iris_df = iris_df[iris_df['target'] != 2] # Remove rows where the value in the "target" column is equal to 2
del iris_df['petal length (cm)'] 
del iris_df['petal width (cm)']
iris_df
```

```python
sns.scatterplot(x = 'sepal length (cm)', y = 'sepal width (cm)', hue = 'target', data = iris_df, style = 'target')
plt.title('Scatter plot of training data')
```
